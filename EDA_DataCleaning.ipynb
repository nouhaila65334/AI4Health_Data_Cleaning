```python
#Data Visualization of the age in function of the charges
import seaborn as sns

sns.set()
plt.figure(figsize = (20,6), dpi = 400)
sns.barplot( x = 'age',  y = 'charges', data = dataset)
```

```python
#Visualizinf data: sex in function of charges

sns.set()
plt.figure(figsize = (20,6), dpi = 400)
sns.barplot( x = 'sex',  y = 'charges', data = dataset)
```

```python
#visualizing data:  region in funciton of charges
sns.set()
plt.figure(figsize = (20,6), dpi = 400)
sns.lineplot( x = 'region',  y = 'charges', data = dataset)
```
